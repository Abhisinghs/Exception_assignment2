```python
#Answer_3
ArithmeticError is simply an error that occurs during numeric calculations.

ArithmeticError types in Python include:

OverFlowError
ZeroDivisionError
FloatingPointError

These errors are all capable of crashing a code in Python. It is essential to
catch an error because you do not want your code to crash as a result of incorrect
input from you or a user.
```

```python
#example of zeroDivisionError
try:
    a=10/0
    print(a)
except ArithmeticError:
    print("you have just made an Arithmetic Error")
```

```python
#example of overflow error
j = 5.0

try:
    for i in range(1, 1000):
        j = j**i
        print(j)
except OverflowError as e:
    print("Overflow error happened")
    print(f"{e}, {e.__class__}")
```

```python
#Answer_4
The LookupError exception in Python forms the base class for all exceptions that are raised
when an index or a key is not found for a sequence or dictionary respectively.

You can use LookupError exception class to handle both IndexError and KeyError exception classes.
- LookupError
 --> IndexError
 --> KeyError
```

```python
#example of handling indexError

#list
list=[1,2,3,4]

try:
    print(list[10])
except LookupError as e:
    print(f"{e},{e.__class__}")
    
#string
name="abhi"

try:
    print(name[15])
except LookupError as e:
    print(f"{e} , {e.__class__}")
    
# tuples
x = (1, 2, 3, 4)
try:
    print(x[10])
except LookupError as e:
    print(f"{e}, {e.__class__}")
```

```python
#example of handling key error

dict={1:"name",2:"roll no."}
try:
    print(dict[3])
except LookupError as e:
    print(f"{e},{e.__class__}")
```
